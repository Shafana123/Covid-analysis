```python
import numpy as np
import scipy.io
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def SIR(Vect,t):
    S,I,R = Vect
    return [-a*S*I, a*S*I - b*I, b*I]
a = 0.1737680
b = 0.34
P = 5.6865E-07
R_0 = 0.0000000140187
I_0 = P - R_0
S_0 = 1-P
Vect_0 = (S_0, I_0, R_0)
```

```python
t = np.linspace(0,800,781)
sol = odeint(SIR, y0 = Vect_0, t = t)
```

```python
P_new = 1-sol
P_new
```

```python
S_sol = sol.T[0]
I_sol = sol.T[1]
R_sol = sol.T[2]
P_new = 1- S_sol
plt.plot(t,S_sol, label = 'S', color = 'Red')
plt.plot(t,I_sol, label = 'I')
plt.plot(t,R_sol, label ='R')
plt.legend(loc = 0)
plt.xlabel("DAYS")
plt.ylabel("fraction of infected population")
plt.title("Brazil - Plot")
plt.show()
```
